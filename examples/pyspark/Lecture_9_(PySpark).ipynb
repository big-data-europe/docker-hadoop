## RDD Creation 

```python
# Import SparkSession
from pyspark.sql import SparkSession

# Create SparkSession 
spark = SparkSession.builder \
      .master("local[1]") \
      .appName("SparkByExamples.com") \
      .getOrCreate() 
```

###using parallelize()

```python
# Create RDD from parallelize    
dataList = [("Java", 20000), ("Python", 100000), ("Scala", 3000)]
rdd=spark.sparkContext.parallelize(dataList)
```

###using textFile()

```python
# Create RDD from external Data source
rdd2 = spark.sparkContext.textFile("sample_data/README.md")
```

```python
rdd2.collect()
```

### using createDataFrame()

```python
data = [('James','','Smith','1991-04-01','M',3000),
  ('Michael','Rose','','2000-05-19','M',4000),
  ('Robert','','Williams','1978-09-05','M',4000),
  ('Maria','Anne','Jones','1967-12-01','F',4000),
  ('Jen','Mary','Brown','1980-02-17','F',-1)
]

columns = ["firstname","middlename","lastname","dob","gender","salary"]
df = spark.createDataFrame(data=data, schema = columns)
```

```python
df.printSchema()
```

```python
df.show() 
```

```python
df = spark.read.option('header','True').csv("zipcodes.csv")
df.printSchema()
# https://github.com/spark-examples/pyspark-examples/tree/master/resources
```

### PySpark SQL Tutorial

```python
df.createOrReplaceTempView("PERSON_DATA")
df2 = spark.sql("SELECT * from PERSON_DATA")
df2.printSchema()
df2.show()
```

```python
groupDF = spark.sql("SELECT city, count(*) from PERSON_DATA group by city")
groupDF.show()
```
